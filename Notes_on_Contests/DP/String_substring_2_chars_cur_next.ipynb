# [Moons and Umbrellas](https://codingcompetitions.withgoogle.com/codejam/round/000000000043580a/00000000006d1145)
- Given
    + String `S`: Only includes 3 types of char `[C, J, ?]`
        + '?' Can be replaced by `C` or `J`
    + `X`: If string S contain a substring `CJ`. Cost += X
    + `Y`: If string S contain a substring `JC`. Cost += Y

- Find: The minimum possible cost

#### Constraints
- $1 \leq \text{len(S)} \leq 1000$
- $-100 \leq X \leq 100$
- $-100 \leq Y \leq 100$

# Examples


#### Sample Inputs/Outputs

- Inputs

```
5
2 3 CJ?CC?
4 2 CJCJ
1 3 C?J
2 5 ??J???
2 -5 ??JJ??
```

- Outputs

```
Case #1: 5
Case #2: 10
Case #3: 1
Case #4: 0
Case #5: -8
```

#### Explanation

```
// Input
X=2 Y=3 S=CJ?CC?

// Output
5

// Explanation
CJ?CC? replaced by CJCCCC or CJJCCC
```


# Solution
- DP based on cur_char, next_char. Total 9 cases
    + cur_char = C, next_char = `[C,J,?]`
    + cur_char = J, next_char = `[C,J,?]`
    + cur_char = ?, next_char = `[C,J,?]`

- Code

```C++
int get_min_cost(const string &S, int X, int Y) {
    int N = S.size();

    // State dp[i][cur_char]: At pos i, cur_char = C or J
    int dp[1005][2];
        // 0: C
        // 1: J
        
    memset(dp, 0x7f, sizeof(dp));

    if(S[0] == 'C' || S[0] == '?') dp[0][0] = 0;
    if(S[0] == 'J' || S[0] == '?') dp[0][1] = 0;

    for(int i=0; i<N-1; ++i) {
        char cur_char = S[i];
        char next_char = S[i+1];

        if(cur_char == 'C' && next_char == 'C') dp[i+1][0] = min(dp[i+1][0], dp[i][0]);
        else if(cur_char == 'C' && next_char == 'J') dp[i+1][1] = min(dp[i+1][1], dp[i][0] + X);
        else if(cur_char == 'C' && next_char == '?') {
            dp[i+1][0] = min(dp[i+1][0], dp[i][0]);           // ? = C
            dp[i+1][1] = min(dp[i+1][1], dp[i][0] + X);       // ? = J
        }

        else if(cur_char == 'J' && next_char == 'C') dp[i+1][0] = min(dp[i+1][0], dp[i][1] + Y);
        else if(cur_char == 'J' && next_char == 'J') dp[i+1][1] = min(dp[i+1][1], dp[i][1]);
        else if(cur_char == 'J' && next_char == '?') {
            dp[i+1][0] = min(dp[i+1][0], dp[i][1] + Y);           // ? = C
            dp[i+1][1] = min(dp[i+1][1], dp[i][1]);               // ? = J
        }

        else if(cur_char == '?' && next_char == 'C') {
            dp[i+1][0] = min(dp[i+1][0], dp[i][0]);     // ? = C
            dp[i+1][0] = min(dp[i+1][0], dp[i][1] + Y); // ? = J
        }
        else if(cur_char == '?' && next_char == 'J') {
            dp[i+1][1] = min(dp[i+1][1], dp[i][0] + X);     // ? = C
            dp[i+1][1] = min(dp[i+1][1], dp[i][1]);         // ? = J
        }
        else if(cur_char == '?' && next_char == '?') {
            dp[i+1][0] = min(dp[i+1][0], dp[i][0]);           // cur = C, next = C
            dp[i+1][1] = min(dp[i+1][1], dp[i][0] + X);       // cur = C, next = J
            dp[i+1][0] = min(dp[i+1][0], dp[i][1] + Y);       // cur = J, next = C
            dp[i+1][1] = min(dp[i+1][1], dp[i][1]);           // cur = J, next = J
        }
    }

    return min(dp[N-1][0], dp[N-1][1]);
}
```

#### Simplify the conditions

```C++
int get_min_cost(const string &S, int X, int Y) {
    int N = S.size();

    // State dp[i][cur_char]: At pos i, cur_char = C or J
    int dp[1005][2];
        // 0: C
        // 1: J
    memset(dp, 0x7f, sizeof(dp));

    if(S[0] != 'J') dp[0][0] = 0;
    if(S[0] != 'C') dp[0][1] = 0;

    for(int i=0; i<N-1; ++i) {
        char cur_char = S[i];
        char next_char = S[i+1];

        if(next_char != 'J') {
            dp[i+1][0] = min(dp[i+1][0], dp[i][0]);
            dp[i+1][0] = min(dp[i+1][0], dp[i][1] + Y);
        }

        if(next_char != 'C') {
            dp[i+1][1] = min(dp[i+1][1], dp[i][0] + X);
            dp[i+1][1] = min(dp[i+1][1], dp[i][1]);
        }
    }

    return min(dp[N-1][0], dp[N-1][1]);
}

```
