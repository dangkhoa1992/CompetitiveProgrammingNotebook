## [Count of Smaller Numbers After Self](https://leetcode.com/problems/count-of-smaller-numbers-after-self/)
- The counts array has the property where counts[i] is the number of smaller elements to the right of nums[i].
- Example

```
Input: [5,2,6,1]
Output: [2,1,1,0] 
Explanation:
    To the right of 5 there are 2 smaller elements (2 and 1).
    To the right of 2 there is only 1 smaller element (1).
    To the right of 6 there is 1 smaller element (1).
    To the right of 1 there is 0 smaller element.
```

#### Solution
- Sort index array to keep the index
- 2nd half of merge sort = right to the 1st half  

```C++
class Solution {
public:
    vector<int> A;
    vector<int> ans;

    void merge(vector<int> &idx, vector<int> &temp, int l, int m, int r) {
        int i = l;
        int j = m + 1;
        int cur = 0;

        int smaller_cnt = 0;
        while(i <= m || j <= r) {
            if(i > m) temp[cur++] = idx[j++];

            // Update counter to 1st half A[i]
            else if(j > r) {
                ans[idx[i]] += smaller_cnt;
                temp[cur++] = idx[i++];
            }

            // Update counter to 1st half A[i]
            else if(A[idx[i]] <= A[idx[j]]) {
                ans[idx[i]] += smaller_cnt;
                temp[cur++] = idx[i++];
            }

            // Prefer 2nd half A[j]: increase counter  
            else {
                ++smaller_cnt;
                temp[cur++] = idx[j++];
            }
        }

        for(int i=0; i < cur; ++i) idx[l + i] = temp[i];
    }
    void mergeSort(vector<int> &idx, vector<int> &temp, int l, int r) {
        if(l >= r) return;

        int m = l+(r-l)/2;
        mergeSort(idx, temp, l, m);
        mergeSort(idx, temp, m+1, r);
        merge(idx, temp, l, m, r);
    }
    vector<int> countSmaller(const vector<int>& nums) {
        A = nums;
        int N = A.size();
        ans.assign(N, 0);

        // Sort index array rather than original array
        vector<int> idx(N); for(int i=0; i<N; ++i) idx[i] = i;
        vector<int> temp(N);
        mergeSort(idx, temp, 0, N-1);

        return ans;
    }
};
```
